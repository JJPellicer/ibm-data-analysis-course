#### 2.4 - Data Normalization in Python.
Uniform the feature value with different range.
#### Methodns of normalizing data.
* **Simple feature scaling:** $$x_{new}=\frac{x_{old}}{x_{max}} $$
* **Min-Max:** $$x_{new}=\frac{x_{old}-x_{min}}{x_{max}-x_{min}}$$
* **Z-score:** $$x_{new}=\frac{x_{old}-\nu}{\sigma}$$
With Pandas:
```python
#Simple feature scaling normalization.
df["lenght"]=df["lenght"]/df["lenght"].max()
#Min max method.
df["lenght"]=(df["lenght"]-df["lenght"].min())/(df["lenght"].max()-df["lenght"].min())
#Z-score method.
df["lenght"]=(df["lenght"]-df["lenght"].mean())/df["lenght"].sdt()
```

### 4.2 - Linear Regression and Multiple Linear Regression.
* Linear regression will refer to one independent variable to make a prediction.
* Multiple Linear Regression  will refer to multiple independent variables to make a prediction.
#### Simple linear regression.
1. The predictor(independent) variable - x.
2. The target (dependent) variable - y.
Linear relationship between the variables: $$ y=b_0+b_1x$$
* $b_0$: The intercept.
* $b_1$: The slope.
We usually store th data points in two data frame or NumPy arrays.
* The value we would like to predict we store in the array y.
* We store the dependent variable in the data frame or array x.
When we have a model we use the hat on the y to denote the model is an estimate. $ \hat{y}=b_0+b_1x$
##### Fitting a simple linear model estimator.
```python
#Import the linear regression.
from sklearn.linear_model import LinearRegression
```

#### Multiple Linear Regression (MLR).
This method is used to explain the relationship between:
* One continuous target (Y) variable.
* Two or more predictor (X) variables.

For example four predictor variables:
$$ \hat{Y}=b_0+b_1x_1+b_2x_2+b_3x_3+b_4x_4$$

### 4.4 - Kernel Density Estimation (KDE) Plots for Model Evaluation.
The Kernel Density Estimation (KDE) Plots are valuable tool for visualizing data distibutions by estimatinf their probability density function (PDF).\
KDE plots are beneficial in model evaluating for the following reasons:
* They provide a smooth approximation of the data distribution.
* They help compare the true vs. predicted distributions effectively.
* Unlike histograms, KDE plots are not sensitive to bin sizes.
* Unlike histograms, KDE plots are not sensitive to bin sizes.
  

`````python
import numpy as npy
import pandas as pds
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
# Generating Sample Data
npy.random.seed(42)
x = npy.random.rand(100) * 10
y = 3 * x + npy.random.normal(0, 3, 100)  # Linear relation with noise
data = pds.DataFrame({'X': x, 'Y': y})
# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(data[['X']], data['Y'], test_size=0.2, random_state=42)
# Training a Model
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
# Plotting KDE for Observed vs. Predicted Values
plt.figure(figsize=(8, 5))
sns.kdeplot(y_test, label='Actual', fill=True, color='blue')
sns.kdeplot(y_pred, label='Predicted', fill=True, color='red')
plt.xlabel('Target Variable')
plt.ylabel('Density')
plt.title('KDE Plot of Actual vs. Predicted Values')
plt.legend()
plt.show()
`````

### 4.6 - Measures for In-Sample Evaluation.
* Measure how well the model fits our data.
* Two important measures to determine the fit of a model:
  * Mean Squared Error (MSE)
  * R-square ($R^2$)
#### Mean Squared Error (MSE)
To measure the MSE, you find the difference between the actual value y and the predicted value y hat then square it.
$$ \frac{\sum^n_{i=1} (y_i-\hat{y_i})^2}{n}$$
In python:
```python
from sklearn.metric import mean_squared_error
mean_squared_error(actual_value,predicted_value)
```
#### R-squared $R^2$
It's a measure to determine how close the data is to the fitted regression line.
* The base model is the average of the y-values.
* $R^2$ compares the regression line to this average.
Coefficient of Determination.
$$R^2=\left(1-\frac{MSE(y,\hat{y})}{MSE(y,\bar{y})}\right)$$
* If $R^2$=1 means that the regression line has a good fit.
* If $R^2$=0 means that the regression line has a poor fit.
To calulate in python:
```python
X=df[["highay-mpg"]]
Y=df["price"]
lm.fit(X,Y)
lm.score(X,Y)
```

### 5.1 - Model Evaluation and Refinement.

* In-sample evaluation tells us how well our model will fit the data used to train it.
* Problem?
  * It does not tell us how the trained model can be used to predict new data.
* Solution?
  * In-sample data or training data.
  * Out-of-sample evaluation or test set.
#### Training/Testing sets.
* Split dataset into:
  * Training set (70%)
  * Testing set (30%)
* Build and train the model with a training set.
* Use testing set to asses the performance of a predictive model.
* When we have completed testing our model we should use all the data to train the model to get the best performance.
#### Function training_test_split().
This function randomly splits a dataset into training and testing subsets.
```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.3,random_state=0)
```
* *x_data:* features or independent variables.
* *y_data:* dataset target.
* *x_train,y_train:* parts of avaliable data as a training set.
* *x_test,y_test:* parts of avaliable data as testing set.
* *test_size:* percentage of the data for testing (here 30%).
* *random_state:* number generator used for random sampling.

#### Generalization perfomance.
* Generalization error is a measure of how well our data does at predicitn previously unseen data.
* The error we obatin using our testing data is an aproximation of this error.
Using a lot of data for training gives us an accurate means of determining how well our model will perform in the real world. But the precision of the performance will be low.

#### Cros validation.
* Most common out-of sample evaluation metrics.
* More effective use of data (each observation is used for both training and testing).
In this method, the dataset is split into k equal groups. Each group is reffered to as a fould. Some of the folds can be used as a training set which we use to train the model, and the remaining parts are used as a test set.\
